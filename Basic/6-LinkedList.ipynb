```python
# Reverse a linked list

class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        prev, curr = None, head

        # None--->1 -> 2 -> 3
        #  ^      ^
        # prev   curr

        while curr:
            temp = curr.next
            curr.next = prev
            prev = curr
            curr = temp

        return prev
```

```python
# Cycle in Linked list
class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        fast, slow = head, head

        while fast and fast.next: ### note here 'and' is there to not over run inside loop for .next.next...
            slow = slow.next
            fast = fast.next.next

            if slow == fast:
                return True
            
        return False
        
```

```python
### Reorder list
# Basically split the linked list in half and revese the second half and then have pointer
# at start and end then merge one from start one form end list that you will reorder in such pattern
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reorderList(self, head: Optional[ListNode]) -> None:
        """
        Do not return anything, modify head in-place instead.
        """
        # Break the list in half using slow and fast pointer
        slow, fast = head, head.next

        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

        second = slow.next
        prev = slow.next = None
        # Reverse the second half of linked list

        while second:
            tmp = second.next
            second.next = prev
            prev = second
            second = tmp

        # Now merge both hslfs

        first, second = head, prev
        while second:
            tmp1, tmp2 = first.next, second.next
            first.next = second
            second.next = tmp1
            first, second = tmp1, tmp2
        
```

```python
# Definition for singly-linked list.
class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        dummy = ListNode()
        cur = dummy

        carry = 0
        while l1 or l2 or carry:
            v1 = l1.val if l1 else 0
            v2 = l2.val if l2 else 0

            # new digit
            val = v1 + v2 + carry
            carry = val // 10
            val = val % 10
            cur.next = ListNode(val)

            # update ptrs
            cur = cur.next
            l1 = l1.next if l1 else None
            l2 = l2.next if l2 else None

        return dummy.next
```

```python
### Find cycle in linked list

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        
        if not head or not head.next:
            return False
        
        slow = head
        fast = head

        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                return True
        
        return False
```

```python
# Floyd's Algo aka turle and hare algo find duplicate in array in constant space
class Solution:
    def findDuplicate(self, nums: List[int]) -> int:
        slow, fast = 0, 0
        while True:
            slow = nums[slow]
            fast = nums[nums[fast]] # this works boz its special array with length equal to 'n' so this will have exactly one duplicate in it
            if slow == fast:
                break

        slow2 = 0
        while True:
            slow = nums[slow]
            slow2 = nums[slow2]
            if slow == slow2:
                return slow
```

```python
### LC - Hard - Merge K sorted list
# Divide and Conquer approach
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def mergeTwoLists(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode(0)
        tail = dummy
        
        while l1 and l2:
            if l1.val < l2.val:
                tail.next = l1
                l1 = l1.next
            else:
                tail.next = l2
                l2 = l2.next
            tail = tail.next
        
        # Append any remaining nodes from l1 or l2
        tail.next = l1 if l1 else l2
        return dummy.next

    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        if not lists:
            return None
        if len(lists) == 1:
            return lists[0]
        
        # Divide and conquer strategy
        def mergeRange(lists, start, end):
            if start == end:
                return lists[start]
            mid = (start + end) // 2
            left = mergeRange(lists, start, mid)
            right = mergeRange(lists, mid + 1, end)
            return self.mergeTwoLists(left, right)
        
        # Call the recursive divide-and-conquer function
        return mergeRange(lists, 0, len(lists) - 1)
```
