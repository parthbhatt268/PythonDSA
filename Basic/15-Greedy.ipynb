```python
### Max sum  of subarray
from typing import List

class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        maxSub = nums[0]
        cursum = 0
        
        for n in nums:
            if cursum < 0:
                cursum = 0
            cursum += n
            maxSub = max(maxSub, cursum)
            
            
        
        return maxSub
```

```python
# Jump Game - we can use DP we will get it in O(n*2) but with greedy we can get it in O(n)
class Solution:
    def canJump(self, nums: List[int]) -> bool:
        goal = len(nums) - 1

        for i in range(len(nums)-1, -1, -1): # yes statting from len - 1 boz we have to consider case when input array is just one element
            if i + nums[i] >= goal:
                goal = i
        
        return True if goal == 0 else False
```

```python
# Max sub array
# Bascially you keep simple rule if you see negavtive restrrt but negative should be total sum until that number then
# you will restart from count zero and thats it
# Solution is O(n)
class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        res = nums[0]

        curSum = 0

        for i in nums:
            if curSum < 0:
                curSum = 0
            curSum += i
            res = max(curSum, res)
        
        return res
```
