```python
# Nice and easy one

class Solution:
    def maxSumDistinctTriplet(self, x: List[int], y: List[int]) -> int:
        pair = {}

        for x, y in zip(x,y):
            if x not in pair or y > pair[x]:
                pair[x] = y


        if len(pair) < 3:
            return -1


        top = heapq.nlargest(3, pair.values())
        return sum(top)
```
