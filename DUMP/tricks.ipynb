```python
from collections import Counter

# Create a Counter object
arr = [1, 2, 2, 3, 3, 3]
freq = Counter(arr)

print(freq)  # Output: Counter({3: 3, 2: 2, 1: 1})

# Access the frequency of an element
print(freq[3])  # Output: 3 (as 3 appears 3 times in the list)
print(freq[2])  # Output: 2 (as 2 appears 2 times in the list)
print(freq[1])  # Output: 1 (as 1 appears 1 time in the list)

# Access an element that is not present
print(freq[5])  # Output: 0 (elements not in the Counter return 0 by default)


# Looping through Counter
for key, value in freq.items():
    print(f"Element: {key}, Frequency: {value}")
```
