```python
### Floyd's ALgorithm aka turle and hare algo
# https://www.youtube.com/watch?v=PvrxZaH_eZ4
## slow, fast pointer bascially the point where the slow and fast intersect then you start
# a second slow pointer which will loop adn they will intersect to whihc the next number 
# will be the duplicate this si mahtematically proven
class Solution:
    def findDuplicate(self, nums: List[int]) -> int:
        slow, fast = 0, 0
        while True:
            slow = nums[slow]
            fast = nums[nums[fast]]
            if slow == fast:
                break

        slow2 = 0
        while True:
            slow = nums[slow]
            slow2 = nums[slow2]
            if slow == slow2:
                return slow
```

```python
### Divide and conquer strategy
# Q. merge K sorted linked lsit
# What approach I thought was like merge first 2 list and then merging their resutl with thirs list
# But this is not optimal approach so what we do is we divide the k list in half till be reacha point
# where we can merge two list and go up by merge them
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def mergeTwoLists(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode(0)
        tail = dummy
        
        while l1 and l2:
            if l1.val < l2.val:
                tail.next = l1
                l1 = l1.next
            else:
                tail.next = l2
                l2 = l2.next
            tail = tail.next
        
        # Append any remaining nodes from l1 or l2
        tail.next = l1 if l1 else l2
        return dummy.next

    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        if not lists:
            return None
        if len(lists) == 1:
            return lists[0]
        
        # Divide and conquer strategy
        def mergeRange(lists, start, end):
            if start == end:
                return lists[start]
            mid = (start + end) // 2
            left = mergeRange(lists, start, mid)
            right = mergeRange(lists, mid + 1, end)
            return self.mergeTwoLists(left, right)
        
        # Call the recursive divide-and-conquer function
        return mergeRange(lists, 0, len(lists) - 1)
```

```python
### Palindrome

class Solution:
    def isPalindrome(self, s: str) -> bool:
        new  = ''
        for i in s:
            if (i.isalpha() or i.isdigit()):
                new += i.lower()
        return (new == new[::-1])
    

        
```

```python
### 2 sum ARR IS I THINK SORTED
#Basically you use 2 pointer as the array is already sorted so you create 2 pointer nad simple checks
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        left, right = 0, len(numbers) - 1
        
        while left < right:
            current_sum = numbers[left] + numbers[right]
            if current_sum == target:
                return [left + 1, right + 1]
            elif current_sum < target:
                left += 1
            else:
                right -= 1
```

```python
### 3 SUM
### Basically you just sort ti first and keep 2 loop first outer loop you just keep one number costant and on rest you just run the other loop with 2 SUm two pointe rtechnique
# and one thing to note is if you find current and prevoius is same then you skip it boz that would lead to duplicate triplet answer.
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        res = []
        nums.sort()

        for i, a in enumerate(nums):
            # Skip positive integers (boz if first number is itsekf postive there is no way we can add any triplet to get zero since there is not neagtive number at first place in sorted array)
            if a > 0:
                break

            if i > 0 and a == nums[i - 1]:
                continue

            l, r = i + 1, len(nums) - 1
            while l < r:
                threeSum = a + nums[l] + nums[r]
                if threeSum > 0:
                    r -= 1
                elif threeSum < 0:
                    l += 1
                else:
                    res.append([a, nums[l], nums[r]])
                    l += 1
                    r -= 1
                    while nums[l] == nums[l - 1] and l < r:
                        l += 1
                        
        return res
```

```python
### Container with max water
#Bascially you keep two pointer one at start ond other at end
# Now in each iertation you claute area if new max found you replace and then make decsion eother 
# left or right to move based on deciding whihc is smaller so that we can find max water
class Solution:
    def maxArea(self, height: List[int]) -> int:
        left = 0
        right = len(height) - 1
        res = 0

        while left < right:
            area = (right - left) * min(height[left], height[right])
            res = max(res, area)
            if height[left] < height[right]:
                left += 1
            else:
                right -= 1

        return res
```

```python
#Best time to buy a stock
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        res = 0
        
        lowest = prices[0]
        for price in prices:
            if price < lowest:
                lowest = price
            res = max(res, price - lowest)
        return res


```

```python
### Rain Water trap
class Solution:
    def trap(self, height: List[int]) -> int:
        if not height:
            return 0

        n = len(height)
        leftMax = [0] * n
        rightMax = [0] * n
        res = 0

        # Fill leftMax array
        leftMax[0] = height[0]
        for i in range(1, n):
            leftMax[i] = max(leftMax[i - 1], height[i])

        # Fill rightMax array
        rightMax[n - 1] = height[n - 1]
        for i in range(n - 2, -1, -1):
            rightMax[i] = max(rightMax[i + 1], height[i])

        # Calculate the trapped water
        for i in range(n):
            res += min(leftMax[i], rightMax[i]) - height[i]

        return res
```

```python
### Sliding window
### Longest substring wihtout repeating
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        charSet = set()
        l = 0
        res = 0

        for r in range(len(s)):
            while s[r] in charSet:
                charSet.remove(s[l])
                l +=1
            charSet.add(s[r])
            res = max(res, len(charSet))

        return  res
```

```python
### Sliding window
# Question - Longest Repeating Character Replacement
# Bacially you increace right pointer of window and match one condition that if the mast ocurring term in a 
# window subtracted from the lengthof winodw is less than K then yes we can still extens but if no then 
# this is the casee that mostly occuring elemet in that window is not not replacemeble by K times so we cannot 
# get longets same char stirng hence we will have to reduce the lenght of string frpm left point moving

class Solution:
    def characterReplacement(self, s: str, k: int) -> int:
        count = {}
        res = 0
        l = 0
        
        for r in range(len(s)):
            count[s[r]] = 1 + count.get(s[r], 0)
            
            # Check if we need to shrink the window
            while (r - l + 1) - max(count.values()) > k:
                count[s[l]] -= 1
                l += 1
                
            
            # Update the result with the maximum window size found so far
            res = max(res, r - l + 1)
        
        return res

```

```python
### Sliding window
# Here fwe are given s1 and s2 we have to return true if there is any permutation of s1 found in s2 else false
# sol - we create a window of s1 size and run that window over s2 and sif any any case the frewuency of window and s1 macthed
# we return true else once we vissted all of s2 then return false
class Solution:
    def checkInclusion(self, s1: str, s2: str) -> bool:
        if len(s1) > len(s2):
            return False
        
        # Create a frequency map for s1
        s1_count = {}
        for char in s1:
            s1_count[char] = s1_count.get(char, 0) + 1
        
        # Create a frequency map for the first window in s2
        window_count = {}
        for i in range(len(s1)):
            char = s2[i]
            window_count[char] = window_count.get(char, 0) + 1
        
        # Check if the first window matches s1_count
        if s1_count == window_count:
            return True
        
        # Slide the window over s2
        for i in range(len(s1), len(s2)):
            start_char = s2[i - len(s1)]
            end_char = s2[i]
            
            # Remove the character that's sliding out of the window
            if window_count[start_char] == 1:
                del window_count[start_char]
            else:
                window_count[start_char] -= 1
            
            # Add the new character that's sliding into the window
            window_count[end_char] = window_count.get(end_char, 0) + 1
            
            # Check if the current window matches s1_count
            if s1_count == window_count:
                return True
        
        # No permutation found
        return False
```

```python
### Minimum window substring
# Here just like above problem you didnt had k as window size given, so the goal of this problem was to find the 
# min k as window size, so you start with one expanding the right and in btw iteration we also do check of map
# and update the min window size if we found smaller one. So we ince in every iteration and simulaneouly check
# and reduce the left as well until right reaches the last point
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        if t == "":
            return ""

        countT, window = {}, {}
        for c in t:
            countT[c] = 1 + countT.get(c, 0)

        have, need = 0, len(countT)
        res, resLen = [-1, -1], float("infinity")
        l = 0
        for r in range(len(s)):
            c = s[r]
            window[c] = 1 + window.get(c, 0)

            if c in countT and window[c] == countT[c]:
                have += 1

            while have == need:
                # update our result
                if (r - l + 1) < resLen:
                    res = [l, r]
                    resLen = r - l + 1
                # pop from the left of our window
                window[s[l]] -= 1
                if s[l] in countT and window[s[l]] < countT[s[l]]:
                    have -= 1
                l += 1
        l, r = res
        return s[l : r + 1] if resLen != float("infinity") else ""
```
